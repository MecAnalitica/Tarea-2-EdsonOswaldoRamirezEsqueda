La aproximación lineal para ángulos pequeños es 

$sin (\theta) \approx \theta$ 

y así la ecuación será una ecuación lineal de segundo orden

$  \ddot \theta + \mu \dot \theta + \omega_0^2 \theta = A cos(\omega_d t) $

Y la solución particular de la expresión es:

$ \theta_p = \frac{A}{\sqrt{(\omega_0^2 - \omega_d^2)^2 + (\mu \omega_d)^2}} cos(\omega_d t - \delta) $

Y sustituyendo los valores dados ($\mu = 0.05$ , $A = 1$ y $ \omega_0 = 1$) se tiene que:

$ \theta_p = \frac{1}{\sqrt{(1 - \omega_d^2)^2 + 0.0025 \omega_d^2}} cos(\omega_d t - \delta) $

Y por consiguiente, la amplitud será:

$ Amplitud = \frac{1}{\sqrt{(1 - \omega_d^2)^2 + 0.0025 \omega_d^2}} $

La cual se gráfica y tiene la siguiente forma:

```python
from pylab import *      

plt.figure(figsize=(10,8))

x = np.arange(0,10,0.001)
y = 1/pow(pow(1 - pow(x,2),2) + pow(0.05*x , 2) , 0.5)
plt.plot(x,y)
pyplot.xlim(-1, 3)
pyplot.ylim(-1, 15)
plt.title("Amplitud contra la fuerza externa")
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
plt.grid()
plt.xlabel("Frecuencia de la fuerza externa $(\omega_d)$",fontsize=18)
plt.ylabel("Amplitud (D)",fontsize=18)

pyplot.savefig("output.png")
plt.show() 
```

Pero para este caso en particular $\omega_0 = 1$ y $\mu = 0.05$

En este caso nos encontramos con un oscilador que por definició es inframortiguado.

```python
from pylab import *      

plt.figure(figsize=(10,8))

x = np.arange(0,100,0.001)
y = np.arctan(0.05*x/(1 - x**2))
plt.plot(x,y)
pyplot.xlim(0, 3)
pyplot.ylim(-1, 1)
plt.title("Amplitud contra la fuerza externa", fontsize=22)
pyplot.axhline(0, color="black")
pyplot.axvline(0, color="black")
plt.grid()
plt.xlabel("Frecuencia de la fuerza externa $(\omega_d)$",fontsize=18)
plt.ylabel("Fase de solucion ($\delta$)",fontsize=18)

pyplot.savefig("output.png")
plt.show() 
```

Para el incisio c) se considera la solución completa de la ecuación diferencial del problema, es decir,

$ \theta = \theta_p + \theta_c $

Donde $\theta_c$ es la solución de la ecuación homogénea, por medio del polinomio característico. 

Entonces, 

$  \ddot \theta + \mu \dot \theta + \omega_0^2 \theta = 0 $

$\theta_c = K e^{\frac{-\mu t}{2}} cos \left[\sqrt{\omega_0^2 - (\mu/2)^2} t \right] $

Por lo tanto, la solución general es:

$\theta = K e^{\frac{-\mu t}{2}} cos \left[\sqrt{\omega_0^2 - (\mu/2)^2} t \right] + \frac{A}{\sqrt{(\omega_0^2 - \omega_d^2)^2 + (\mu \omega_d)^2}} cos(\omega_d t - \delta) $

2.- Ecuación no lineal no forzada sin amortiguamiento.

a) Para un ángulo inicial de 70 grados en reposo y una frecuencia natural unitaria, realiza una gráfica donde compares la solucion numérica del ángulo y su velocidad con aquella de la ecuación lineal usando las mismas condiciones iniciales.

La ecuación no lineal no forzada sin amortiguamiento implica que $\mu = 0$  y que no hay una fuerza externa.

$ \ddot \theta + \omega_0^2 sen \theta $

La condición inicial se reescribe en radianes, $\theta (0) = \theta_0 = \frac{7 \pi}{18} $

La solución lineal con esas condiciones se obtiene a continuación:

$sin (\theta) \approx \theta$ 

$\theta = \frac{7 \pi}{18} cos(t)$

Luego,

$\theta =  \frac{-7 \pi}{18} sen(t)$

A continuación se muestran las gráficas:


```python
import matplotlib. pyplot as plt
from matplotlib import pyplot
import numpy as np
import math
from scipy.integrate import odeint

def pend(y, t):
    theta, omega = y
    dydt = [omega, -np.sin(theta)]
    return dydt


y0 = [7*np.pi/18,0]

t = np.linspace(0, 20, 1000)


sol = odeint(pend, y0, t)
plt.plot(t, sol[:, 0], 'b', label='Angulo(solucion numerica)')
plt.plot(t, (7*np.pi/18)*np.cos(t), 'g', label='Angulo(aproximacion lineal)')
plt.title("Angulo contra tiempo", fontsize=22)
plt.legend(loc='best')
plt.xlabel('tiempo')
plt.ylabel('Angulo')
plt.grid()

plt.show()
```

3.- Ecuación no lineal amortiguada no forzada.

a) Utiliza un parametro de amortiguamiento de 0.02, la frecuencia natural unitaria y observa cómo el espacio fase se distorciona. Compara cada trayectoria del inciso b) del ejercio anterior con este caso amortiguado.

Entonces la ecuación que se obtiene es:

$ \ddot \omega + 0.02 \dot \omega + sin \theta = 0$

Y su diagrama de fases es:

```python
%matplotlib inline
import math
import numpy as np                                       
import matplotlib.pyplot as plt                          
from scipy.integrate import odeint


def pend(y,t):
    theta, omega = y
    dydt = [omega, -0.02*omega - np.sin(theta)]
    return dydt

y0 = [4*np.pi,1]

t = np.linspace(0, 20, 1000)

sol = odeint(pend, y0, t)
plt.plot(sol[:, 0] , sol[:, 1], 'b')
plt.title("Espacio fase", fontsize=22)
plt.legend(loc='best')
plt.xlabel('Angulo')
plt.ylabel('Velocidad angular')
plt.grid()

plt.show()  
```

En el inciso b del ejercicio anterior se obtuvieron espacios de fase que forman trayectorias cerradas, mientras que en este las trayectorias son abiertas.
